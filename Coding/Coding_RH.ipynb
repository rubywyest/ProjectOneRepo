```python
import pandas as pd
import numpy as np
import os
import csv
import matplotlib.pyplot as plt
```

```python
in_states = os.path.join("..","Resources","BLS_state_M2018_dl.csv")
states_df = pd.read_csv(in_states)

in_nation = os.path.join("..","Resources","BLS national_M2018_dl.csv")
nation_df = pd.read_csv(in_nation)
```

```python
#Clean Data for field Total Employee count field for use

#convert null data represented by "**' to zeros
nation_df.loc[states_df["TOT_EMP"] == '**', "TOT_EMP"] = "0"
states_df.loc[states_df["TOT_EMP"] == '**', "TOT_EMP"] = "0"

#convert Total column to integer
nation_df['TOT_EMP'] = nation_df['TOT_EMP'].str.replace(',', '').astype(int)
states_df['TOT_EMP'] = states_df['TOT_EMP'].str.replace(',', '').astype(int)
```

```python
#Major occupations per state 
states_major_df = states_df.loc[states_df["OCC_GROUP"] == "major", :].reset_index(drop="true")

states_major_df.sort_values("TOT_EMP")
```

```python
#Major IT Occupations by state

states_major_df = states_df.loc [states_df["OCC_CODE"].str.contains("15-"), :].reset_index(drop = "true")
states_major_df
                                                                                          
```

```python
#Group by State
states_group = states_major_df.groupby(['STATE'])
```

```python
#Consolidating States by "Total employees"
states_major = states_group ['TOT_EMP'].sum()
states_major_df = pd.DataFrame(states_major).reset_index()

#sort by "Total employees"
states_major_sort = states_major_df[[ 'STATE','TOT_EMP',]].sort_values("TOT_EMP", ascending = False).reset_index(drop = True)


#top 10 states
states_major_top = states_major_sort.head(10)
states_major_top
```

```python
# Plot Horizontal bar graph
fig, ax = plt.subplots(figsize = (12,9))

x_axis_states = states_major_top['TOT_EMP'].tolist()
y_axis_states = states_major_top['STATE'].tolist()

height_y = .8
rects1 = ax.bar(17 , y_axis_states, color = "black")

ax.barh(y_axis_states, x_axis_states,color = 'green', height = height_y, align = 'center')
ax.set_yticks(y_axis_states)
ax.set_yticklabels(y_axis_states)
ax.invert_yaxis()

ax.set_xticks(np.arange( 0, max(x_axis_states)+5, 250000))
ax.set_xticklabels(np.arange( 0, max(x_axis_states)+5, 250000))

ax.set_xlabel("TOTAL EMPLOYEES", fontsize = 16)
ax.set_title("EMPLOYEES PER STATE-2018", fontsize = 20)
ax.set_ylabel("STATES", fontsize = 16)
plt.tight_layout()


def autolabel(rects):
    for ii,rect in enumerate(rects):
        height = rect.get_height()
        wide = rect.get_width()
        label = "{:,}".format(x_axis_states[ii])
        ax.text(x_axis_states[ii] + 16, 1.02*height, label) 
      
autolabel(rects1)

# Save the Figure
plt.savefig("../Output_charts/Employees_State.png", bbox_inches = "tight")
```

```python
#Detail IT jobs by state
states_detail_df = states_df.loc [states_df["OCC_CODE"].str.contains("15-"), :].reset_index(drop = "true")

states_detail_df
```

```python
#detail IT jobs in 'TX'
states_detail_tx1 = states_detail_df.loc[states_detail_df['ST'] == 'TX' , : ].reset_index()

#drop the sumarize row
states_detail_tx = states_detail_tx1.drop([0])
states_detail_tx.head()

```

```python
# Plot -Horizontal bar graph
fig, ax = plt.subplots(figsize = (12,10))

x_axis_states = states_detail_tx['TOT_EMP'].tolist()
y_axis_states = states_detail_tx['OCC_TITLE'].tolist()

height_y = .8
rects1 = ax.bar(50 , y_axis_states, color = "red")

ax.barh(y_axis_states, x_axis_states,color = 'green', height = height_y, align = 'center')
ax.set_yticks(y_axis_states)
ax.set_yticklabels(y_axis_states)
ax.invert_yaxis()

ax.set_xticks(np.arange(0, max(x_axis_states)+5, 10000))
ax.set_xticklabels(np.arange( 0, max(x_axis_states)+5, 10000))

ax.set_xlabel("TOTAL EMPLOYEES", fontsize = 16)
ax.set_title("IT JOBS IN TEXAS 2018", fontsize = 20)
ax.set_ylabel("JOB TITLE", fontsize = 16)
plt.tight_layout()

def autolabel(rects):
    for ii,rect in enumerate(rects):
        height = rect.get_height()
        wide = rect.get_width()
        label = "{:,}".format(x_axis_states[ii])
        ax.text(x_axis_states[ii] + 16, 1.02*height, label) 
      
autolabel(rects1)
# Save the Figure
plt.savefig("../Output_charts/IT_jobs_tx.png", bbox_inches = "tight")
```
